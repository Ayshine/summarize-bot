```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text2text-generation", model="musabg/mt5-xl-tr-summarization")

# Load model directly
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("musabg/mt5-xl-tr-summarization")
model = AutoModelForSeq2SeqLM.from_pretrained("musabg/mt5-xl-tr-summarization")

# Read the text from the file
with open('/content/alperRUST.txt', 'r') as file:
    text = file.read()

# Encode the text and generate the summary
inputs = tokenizer.encode(text, return_tensors='pt')
outputs = model.generate(inputs, max_length=1000, min_length=100, length_penalty=2.0, num_beams=4, early_stopping=True)
summary = tokenizer.decode(outputs[0])

# Write the summary to a new file
with open('input-summary.txt', 'w') as file:
    file.write(summary)
```
